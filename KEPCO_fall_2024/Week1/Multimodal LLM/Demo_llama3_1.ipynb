```python
from unsloth import FastLanguageModel
import torch
```

```python
max_seq_length = 2048
dtype = None
load_in_4bit = True

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Meta-Llama-3.1-8B-Instruct-bnb-4bit",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
)
```

```python
FastLanguageModel.for_inference(model)
```

```python
alpaca_prompt = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

### Instruction:
{}

### Input:
{}

### Response:
{}"""
```

```python
instruction = "H√£y vi·∫øt m·ªôt b√†i blog ng·∫Øn v·ªÅ l·ª£i √≠ch c·ªßa vi·ªác ƒë·ªçc s√°ch."
input = "Ti√™u ƒë·ªÅ: L·ª£i √≠ch c·ªßa vi·ªác ƒë·ªçc s√°ch M√¥ t·∫£: B√†i blog ng·∫Øn n√†y s·∫Ω gi·∫£i th√≠ch nh·ªØng l·ª£i √≠ch m√† vi·ªác ƒë·ªçc s√°ch mang l·∫°i cho con ng∆∞·ªùi."

inputs = tokenizer(
[
    alpaca_prompt.format(
        f"{instruction}", # instruction
        f"{input}", # input
        "", # output - leave this blank for generation!
    )
], return_tensors = "pt").to("cuda")

from transformers import TextStreamer
text_streamer = TextStreamer(tokenizer)
_ = model.generate(**inputs, streamer = text_streamer, max_new_tokens = 512)
```
